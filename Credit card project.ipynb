```python
#Scaling the numerical columns

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

### Performing Rule Based Detection

```python
# Rule 1: Flag transactions with an unusually high amount
rule1 = df['Amount'] > df['Amount'].mean() + 5*df['Amount'].std()

# Explanation:
# Calculating the mean and standard deviation of the 'Amount' column.
# Any transaction higher than mean + 5*std is extremely unusual (statistical outlier).
# This rule flags transactions that are suspiciously large.

# Rule 2: Flag transactions that are unusually fast
rule2 = df['transaction_speed'] > df['transaction_speed'].mean() + 4*df['transaction_speed'].std()

# Explanation:
# 'transaction_speed' measures how fast a transaction occurs.
# Extremely fast transactions could indicate automated/fraudulent behavior.
# Here we flag transactions faster than mean + 4*std as suspicious.

# Rule 3: Flag accounts with unusually high activity
rule3 = df['account_activity_score'] > df['account_activity_score'].mean() + 4*df['account_activity_score'].std()

# Explanation:
# 'account_activity_score' shows how active a customer’s account is.
# Abnormally high activity may suggest suspicious patterns.
# We flag scores significantly above the average.

# Rule 4: Flag transactions with unusually high geo_distance
rule4 = df['geo_distance'] > df['geo_distance'].mean() + 4*df['geo_distance'].std()

# Explanation:
# 'geo_distance' measures distance between transaction location and customer’s home/normal location.
# Very high distance could indicate unusual or fraudulent behavior.
# Transactions above mean + 4*std are flagged.

# Combine all rules into a single 'rule_score'
df['rule_score'] = rule1.astype(int) + rule2.astype(int) + rule3.astype(int) + rule4.astype(int)

# Explanation:
# Each rule returns True/False (1/0). 
# Adding them up gives a score: higher score = more rules triggered = more suspicious.

# Create a final flag based on rule_score
df['rule_flag'] = df['rule_score'] > 0

# Explanation:
# If at least one rule is triggered, the transaction is flagged as suspicious.
# You can adjust this threshold depending on how strict you want your rules to be.
```

### Unsupervised Anomaly Detection

```python
df
```

```python
#Unsupervised Anomaly Detection using IsolationForest

from sklearn.ensemble import IsolationForest

# Initialize IsolationForest
iso = IsolationForest(contamination=0.001, random_state=40)

# Fit on numeric features
iso.fit(df[numeric_features])

# Get anomaly scores (higher = more anomalous)
df["anomaly_score"] = iso.decision_function(df[numeric_features])

# Flag anomalies based on score
df["anomaly_flag"] = iso.predict(df[numeric_features])
df["anomaly_flag"] = df["anomaly_flag"].map({1: False, -1: True})
```

```python
#saving the model
import joblib
joblib.dump(iso, "isolation_forest_model.pkl")
```

```python
df
```

### Combine Rule + Anomaly Scores

```python
# Normilizing the rule score

df["rule_score_normalized"] = df["rule_score"] / df["rule_score"].max()
```

```python
# Computing the Hybrid Score

alpha = 0.5
df["final_score"] = alpha * df["rule_score_normalized"] + (1-alpha) * df["anomaly_score"]
df["flag"] = df["final_score"] > 0.5
```
